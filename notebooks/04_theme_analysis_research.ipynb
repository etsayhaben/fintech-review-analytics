```python
import pandas as pd
from config.config import ANALYSIS_CSV_PATH
df = pd.read_csv(ANALYSIS_CSV_PATH)

# Count themes per bank
theme_counts = df.groupby(['bank', 'theme']).size().unstack(fill_value=0)
print(theme_counts)

# Visualize
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(x='theme', hue='bank', data=df)
plt.xticks(rotation=45)
plt.title("Theme Distribution by Bank")
plt.show()

# Optional: top keywords per theme
keywords_per_theme = df.groupby('theme')['cleaned'].apply(lambda x: " ".join(x))
# Could do TF-IDF extraction here for demonstration
```
