```python
import pandas as pd
from config.config import ANALYSIS_CSV_PATH
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(ANALYSIS_CSV_PATH)



# If sentiment_score doesn't exist, create it from sentiment_label
if 'sentiment_score' not in df.columns:
    sentiment_map = {'negative': -1, 'neutral': 0, 'positive': 1}
    df['sentiment_score'] = df['sentiment'].map(sentiment_map)

# Count sentiment labels per bank
sentiment_counts = df.groupby(['bank', 'sentiment']).size().unstack(fill_value=0)
print("Sentiment counts per bank:\n", sentiment_counts)

# Mean sentiment score per bank
mean_scores = df.groupby('bank')['sentiment_score'].mean()
print("Mean sentiment score per bank:\n", mean_scores)

# Quick plot
plt.figure(figsize=(8,6))
sns.countplot(x='sentiment', hue='bank', data=df)
plt.title("Sentiment Distribution by Bank")
plt.show()
```
